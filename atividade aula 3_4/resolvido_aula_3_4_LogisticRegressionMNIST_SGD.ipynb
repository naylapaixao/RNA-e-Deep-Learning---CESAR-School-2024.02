# Regressão Softmax com dados do MNIST utilizando gradiente descendente estocástico por minibatches

O objetivo deste notebook é ilustrar
- o uso do gradiente estocástico por mini-batchs
- utilizando as classes Dataset e DataLoater.

A apresentação da perda nos gráficos é um pouco diferente da usual, mostrando a perda de cada um dos vários minibatches dentro de cada época, de forma que as épocas são apresentadas com valores fracionários.

## Importação das bibliotecas

```python
%matplotlib inline
import matplotlib as plt
import numpy as np
import matplotlib.pyplot as plt

import torch
from torch.autograd import Variable
from torch.utils.data import DataLoader

import torchvision
from torchvision.datasets import MNIST
```

## Dataset e dataloader

### Definição do tamanho do minibatch

```python
batch_size = 100
```

### Carregamento, criação dataset e do dataloader

```python
dataset_dir = 'MNIST/'

dataset_train = MNIST(dataset_dir, train=True, download=True,
                      transform=torchvision.transforms.ToTensor())

loader_train = DataLoader(dataset_train, batch_size=batch_size, shuffle=True)

print('Número de minibatches de trenamento:', len(loader_train))

x_train, y_train = next(iter(loader_train))
print("\nDimensões dos dados de um minibatch:", x_train.size())
print("Valores mínimo e máximo dos pixels: ", torch.min(x_train), torch.max(x_train))
print("Tipo dos dados das imagens:         ", type(x_train))
print("Tipo das classes das imagens:       ", type(y_train))
```

### Usando todas as amostras do MNIST

Neste exemplo utilizaremos todas as amostras de treinamento.

```python
print('Número de minibatches de trenamento:', len(loader_train))
n_batches_train = len(loader_train)
total_samples = dataset_train.train_data.size(0)
```

## Modelo

```python
model = torch.nn.Linear(28*28, 10)
```

```python
x = torch.ones(28*28).reshape(1, 784)
print(x.shape)
predict = model(x)
predict
```

## Treinamento

### Inicialização dos parâmetros

```python
n_epochs = 5
learningRate = 0.5

# Utilizaremos CrossEntropyLoss como função de perda
criterion = torch.nn.CrossEntropyLoss()

# Gradiente descendente
optimizer = torch.optim.SGD(model.parameters(), lr=learningRate)
```

### Laço de treinamento dos parâmetros

```python
epochs = []
losses = []

total_trained_samples = 0
for i in range(n_epochs):
    for k,(x_train, y_train) in enumerate(loader_train):
        # Transforma a entrada para uma dimensão
        inputs = Variable(x_train.view(-1, 28 * 28))
        # predict da rede
        outputs = model(inputs)

        # calcula a perda
        loss = criterion(outputs, Variable(y_train))

        # zero, backpropagation, ajusta parâmetros pelo gradiente descendente
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_trained_samples += x_train.size(0)
        epochs.append(total_trained_samples / total_samples)
        losses.append(loss.data)
```

```python
print('Final loss:', loss.data)
```

### Visualizando gráfico de perda durante o treinamento

```python
plt.plot(epochs, losses)
plt.xlabel('época')
```

### Visualização usual da perda, somente no final de cada minibatch

```python
plt.plot(epochs[:5 * n_batches_train +1:n_batches_train], losses[:5 * n_batches_train+1:n_batches_train])
plt.xlabel('época')
```

2. Altere o tamanho de minibatch (batch_size) algumas vezes, refaça o treinamento, e compare no gráfico abaixo a queda da perda para cada tamanho de minibatch.

```python
dataset_dir = 'MNIST/'

dataset_train = MNIST(dataset_dir, train=True, download=True,
                      transform=torchvision.transforms.ToTensor())
learningRate = 0.5

def train_model(batch_size, n_epochs=5):
    #Carregar os dados
    loader_train = DataLoader(dataset_train, batch_size=batch_size, shuffle=True)
    n_batches_train = len(loader_train)

    #Iniciar os parametros do modelo
    model = torch.nn.Linear(28*28, 10)
    criterion = torch.nn.CrossEntropyLoss()
    optimizer = torch.optim.SGD(model.parameters(), lr=learningRate)

    #Treinamento
    epochs = []
    losses = []

    total_trained_samples = 0
    total_samples = len(dataset_train)
    for i in range(n_epochs):
        for k,(x_train, y_train) in enumerate(loader_train):
            # Transforma a entrada para uma dimensão
            inputs = Variable(x_train.view(-1, 28 * 28))
            # predict da rede
            outputs = model(inputs)

            # calcula a perda
            loss = criterion(outputs, Variable(y_train))

            # zero, backpropagation, ajusta parâmetros pelo gradiente descendente
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()

            total_trained_samples += x_train.size(0)
            epochs.append(total_trained_samples / total_samples)
            losses.append(loss.data)

    return epochs, losses
```

```python
batch_sizes = [32, 100, 500]

plt.figure(figsize=(12, 6))

for bs in batch_sizes:
    print(f"Treinando com batch_size = {bs}")
    epochs, losses = train_model(bs, n_epochs=5)

    # Plotar a perda por época fracionária
    plt.plot(epochs, losses, label=f'batch_size={bs}')

plt.xlabel('Época')
plt.ylabel('Perda')
plt.title('Comparação da Queda da Perda para Diferentes Tamanhos de Minibatch')
plt.legend()
plt.grid(True)
plt.show()
```

## Conclusões sobre os experimentos deste notebook

- Batch sizes menores o ruído é maior, com maior variação na perda;
- Batch sizes maior haverá menor ruído, e a queda da perda é mais suave e estável.
